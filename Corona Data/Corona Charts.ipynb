```python
florida = states[states.state == 'Florida']
fl = florida.reset_index(drop=True)
fl
```

```python
cal = states[states.state == 'California']
cal = cal.reset_index(drop = True)
cal
```

```python
tex = states[states.state == 'Texas']
tex = tex.reset_index(drop = True)
tex
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(nym.cases/1000)
ax.plot(fl.cases / 1000)
plt.title('total cases from first reported case-07/19/20')
plt.ylabel('thousands of cases')
plt.xlabel('days since first case')
plt.legend(['New York','Florida'])
```

```python
present = states[states.date == '2020-07-19']
present
```

```python
present_states = present.drop([present.index[56],present.index[55],present.index[49],present.index[41],present.index[36],present.index[8],present.index[11]])

present_states
```

```python
states_only =present_states.reset_index(drop=True)
states_only
```

```python
less = present[present.cases < 70000]
more = present[present.cases > 70000]
more
```

```python
other_cases = sum(less.cases)
other_deaths = sum(less.deaths)
other_deaths
```

```python
other_row = {'state': 'Other', 'cases': other_cases, 'deaths': other_deaths}
more_other = more.append(other_row,ignore_index=True)
more_other
```

```python
population = pd.read_csv('population.csv')
population
```

```python
population_states = population.drop([population.index[0],population.index[9],population.index[52]])
population_states 
```

```python
population_states = population_states.reset_index(drop=True)
population_states
```

```python
plt.pie(more_other.cases,labels=more_other.state)
plt.title('Spread of Total US Cases as of 07/19/20')
```

```python

plt.scatter(population_states.POPESTIMATE2019,states_only.cases)
plt.subplots_adjust(right = 2)
```

```python

plt.subplot(2,2,1)
plt.plot(us.cases / 1000)
plt.plot(us.deaths/ 1000)
plt.title('US Coronavirus cases and deaths from 01/21/20-07/19/20')
plt.ylabel('thousands of cases/deaths')
plt.xlabel('days since first case')
plt.legend(['cases','deaths'])

plt.subplot(2,2,2)
plt.pie(more_other.cases,labels=more_other.state)
plt.title('Spread of Total US Cases as of 07/19/20')

plt.subplot(2,2,3)
pl = plt.scatter(population_states.POPESTIMATE2019/1000000,states_only.cases/1000)
plt.title('Coronavirus Cases VS Population')
plt.xlabel('population(millions)')
plt.ylabel('thousands of cases')

plt.subplot(2,2,4)
plt.plot(nym.cases/1000)
plt.plot(fl.cases/1000)
plt.plot(cal.cases/1000)
plt.plot(tex.cases/1000)
plt.title('US Coronavirus cases by state from date of first case until 07/19/20')
plt.ylabel('thousands of cases')
plt.xlabel('days since first case')
plt.legend(['New York','Florida','California','Texas'])

plt.subplots_adjust(right = 2)
plt.subplots_adjust(top = 2)



plt.show()
```
